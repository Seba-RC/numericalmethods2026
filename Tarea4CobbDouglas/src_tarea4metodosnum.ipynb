## Tarea 4 - SENL Cobb Douglas

En esta tarea, debemos solucionar una función Cobb-Douglas intentando varios métodos y parámetros.

```python
import numpy as np
from scipy.optimize import root

# parámetros
A1, A2 = 1.2, 0.9
alpha1, alpha2 = 0.3, 0.6
Kbar, Lbar = 100.0, 50.0

def F(x):
    K1, L1, r, w = x
    K2 = Kbar - K1
    L2 = Lbar - L1
    
    # ecuaciones de equilibrio
    f1 = r - A1*alpha1*(np.sign(K1) * (np.abs(K1))**(alpha1-1))*(np.sign(L1) * (np.abs(L1))**(1-alpha1))
    f2 = r - A2*alpha2*(K2**(alpha2-1))*(L2**(1-alpha2))
    f3 = w - 1*A1*(1-alpha1)*(K1**alpha1)*(L1**(-alpha1))
    f4 = w - A2*(1-alpha2)*(K2**alpha2)*(L2**(-alpha2))
    
    return [f1, f2, f3, f4]

# condición inicial
x0 = [50, 25, 0.5, 0.5]

# solucion del sistema no lineal
sol = root(F, x0, method='hybr')
#sol = root(F, x0)

print(sol)
print("---------------")
print("K1 = %.6f, L1 = %.6f, r = %.6f, w = %.6f" % tuple(sol.x))

# capital y trabajo del segundo sector
K2, L2 = Kbar - sol.x[0], Lbar - sol.x[1]
print("K2 = %.6f, L2 = %.6f" % (K2, L2))


```

La principal pregunta es... como podemos encontrar parámetros que nos den una solución?

Pensemos lógicamente en esto, y saquemos algunas condiciones...

- r, el cambio del costo de capital, no puede ser negativo
- w, el cambio del salario del trabajo, tampoco puede ser negativo
- La produccion en si, Y, para cada sector, no puede ser negativa
- Lo mismo se repite para todas las otras variables, L, W, alpha y beta

Entonces sabemos que ninguna variable puede ser negativa.

Probemos entonces, otros parametros iniciales.

Sabemos que ambos sectores son diferentes, debido a que $ \alpha_1 $ y $ \alpha_2 $ son diferentes. Dado esto, podemos modificar los valores iniciales de L1 y K1, y derivar los valores iniciales CORRECTOS de r y w usando estos. 

Asumiendo un equilibrio...

Con un alpha de 0.3, se puede decir que el sector 1 necesita menos capital (K1) que el sector 2. Asignemosle menos al sector 1..
Con un alpha de 0.6, se puede decir que el sector 1 es mejor usando la mano de obra (L1) que el sector 2. Le asignaremos mas al sector 1 

Por esto, los valores L1 y K1, o los valores de r y w, deberian cambiar. Asignar valores iguales en terminos de proporcion para cada uno seria el principal error.

Demosle una proporcion de 70% a L1 en Lbar, y 30% a K1 en Kbar. Con esto, $L_1 = 35$ y $K1 = 30$ 



```python
# Con A1 = 1.2, a1 = 0.3, K1 = 30, L1 = 35,  
A1 = 1.2
alpha1 = 0.3
K1 = 30
L1 = 35

r = A1* alpha1 * (L1 ** (alpha1 - 1)) * (K1 ** (1 - alpha1))
w = A1* (1-alpha1) * (L1 ** (alpha1 - 1)) * (K1 ** (- alpha1))
x0 = [L1, K1, r, w]
x0
```

Ahora que sabemos como funciona la funcion, volvamos a los parametros originales y la solucion simple que encontramos y hagamos pruebas de los distintos metodos para encontrar raices.

```python
# parámetros
A1, A2 = 1.2, 0.9
alpha1, alpha2 = 0.3, 0.6
Kbar, Lbar = 100.0, 50.0

def F(x):
    K1, L1, r, w = x
    K2 = Kbar - K1
    L2 = Lbar - L1
    
    # ecuaciones de equilibrio
    f1 = r - A1*alpha1*(np.sign(K1) * (np.abs(K1))**(alpha1-1))*(np.sign(L1) * (np.abs(L1))**(1-alpha1))
    f2 = r - A2*alpha2*(K2**(alpha2-1))*(L2**(1-alpha2))
    f3 = w - 1*A1*(1-alpha1)*(K1**alpha1)*(L1**(-alpha1))
    f4 = w - A2*(1-alpha2)*(K2**alpha2)*(L2**(-alpha2))
    
    return [f1, f2, f3, f4]

K1 = 30
L1 = 35

r = A1* alpha1 * (L1 ** (alpha1 - 1)) * (K1 ** (1 - alpha1))
w = A1* (1-alpha1) * (L1 ** (alpha1 - 1)) * (K1 ** (- alpha1))
x0 = [L1, K1, r, w]
```

```python
K1List = []
L1List = []
rList = []
wList = []
timeTakenList = []
statusList = []
L2List = []
K2List = []
Method = ['hybr','lm','broyden1','broyden2','anderson','linearmixing','excitingmixing','krylov','df-sane'] # diagbroyden simplemente no funciona con estos valores

for method in Method:
    timeStart = time()
    try:
        sol = root(F, x0, method=method)
        timeEnd = time()
        timeTaken = timeEnd - timeStart
        K1List.append(sol.x[0])
        L1List.append(sol.x[1])
        rList.append(sol.x[2])
        wList.append(sol.x[3])
        timeTakenList.append(timeTaken)
        statusList.append(sol.success)
        K2List.append(Kbar - sol.x[0])
        L2List.append(Lbar - sol.x[1])
    except ValueError:
        sol = root(F, x0, method=method)
        timeEnd = time()
        timeTaken = timeEnd - timeStart
        K1List.append('Failed')
        L1List.append('Failed')
        rList.append('Failed')
        wList.append('Failed')
        timeTakenList.append(timeTaken)
        statusList.append('Failed')
        K2List.append('Failed')
        L2List.append('Failed')

x0forDf = [x0] * len(Method)

print(x0forDf)

dfResults3 = pd.DataFrame(data={
    'Semilla': x0forDf,
    'K1': K1List,
    'L1': L1List,
    'r': rList,
    'w': wList,
    'timeTaken': timeTakenList,
    'status': statusList,
    'K2': K2List,
    'L2': L2List,
    'Metodo': Method
})

display(dfResults3)


```
